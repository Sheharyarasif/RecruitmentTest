```python
class Node:
    
    def __init__(self,val=0, nextval=None):
        self.val = val
        self.nextval =nextval 

head=Node('mon')
n1=Node('tues')
n2=Node('wed')
n3=Node('thurs')
head.nextval=n1
n1.nextval=n2
n2.nextval=n3
class solution:
    def __init__(self,head=None):
        self.pointer=head
    
    def listprint(self):
        printval = self.pointer
        
        while printval is not None:
            
            print (printval.val)
            printval = printval.nextval
    
    
    def sol(self,n,head=None):
        count=self.pointer
        size=0
        while count is not None:
            count=count.nextval
            size+=1
        print('size is ',size)
        
        drop=size-n
        print('one to be dropped is',drop)
        
        count=count1=self.pointer
        
       
        if(size==1):#only one element
            self.pointer=head=None
            return head
        
        elif(n==1):#dropping last
            count1=count.nextval
            while count1.nextval is not None:
                count=count.nextval
                count1=count1.nextval
                
            count.nextval=None
            
        elif(n==size):#dropping first
            print('dropping first')
            self.pointer=head.nextval
            head=None
            head=self.pointer
            
    
        else:#other scenarios
            count1=count.nextval
            i=1
            while i < drop:
                count=count.nextval
                count1=count1.nextval
                i+=1
           
            count.nextval=count1.nextval
            count1=None
        self.listprint()
            
            

        
        
        

d=solution(head)
d.listprint()
d.sol(3,head)
```
